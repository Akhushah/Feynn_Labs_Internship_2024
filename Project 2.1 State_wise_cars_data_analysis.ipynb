```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
```

```python
data=pd.read_csv("/content/states_data_car.csv")
data.head(20)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
sns.countplot(x=data["subsidy"])
plt.title('Count plot of subsidy')
plt.xlabel('Subsidy')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x=data["road tax"])
plt.title('Count plot of road tax')
plt.xlabel('road tax')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming 'data' is your DataFrame containing columns 'state' and 'petrol'

plt.figure(figsize=(20, 5))
sns.set_theme(style="whitegrid")
sns.barplot(x='state', y='petrol', data=data, color='lightslategrey')
plt.title('Petrol Price in Different States', fontsize=16)
plt.ylabel('Petrol Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.set_theme(style="whitegrid")
sns.barplot(x='state', y='diesel', data=data, color = 'lightslategrey')
plt.title('Diesel Price at different states' , fontsize = 16)
plt.ylabel('Diesel Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```
