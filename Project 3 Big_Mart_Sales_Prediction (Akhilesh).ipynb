### Outlier

```python
# Item_Outlet_Sales to see Outiler using Box Plot

plt.figure(figsize = (5,3))
sns.boxplot(x = data['Item_Outlet_Sales'], palette = 'BuPu')
plt.title('Item Outlet Sales Distribution')
```

```python
# Removing Outilers

def outliers(df, features):
    q1 = df[features].quantile(0.25)
    q3 = df[features].quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    return upper_limit, lower_limit

up, low = outliers(data, "Item_Outlet_Sales")
print('Upper Whisker : ', up)
print('Lower Whisker : ', low)

data = data[(data['Item_Outlet_Sales'] > low) & (data['Item_Outlet_Sales'] < up)]
```

```python
# Item_Outlet_Sales after removing Outiler using Box Plot

plt.figure(figsize = (5,3))
sns.boxplot(x = data['Item_Outlet_Sales'], palette = 'BuPu')
plt.title('Item Outlet Sales Distribution after removing outliers')
```

```python
# Change Establishment_Year to Outlet_Age

data['Outlet_Age'] = 2013 - data['Outlet_Establishment_Year']
```

```python
# Consider 0 as missing value in Item_visibility

print('Missing Values in Item_Visibility: ', len(data[num_data['Item_Visibility'] == 0]))

miss_values = (data['Item_Visibility'] == 0)
item_vis_avg = data.pivot_table(values='Item_Visibility', index ='Item_Identifier')

data.loc[miss_values, 'Item_Visibility'] = data.loc[miss_values, 'Item_Identifier'].apply(lambda x:item_vis_avg.loc[x])
print('Missing values after filling: ' , data['Item_Visibility'].isnull().sum())
```

## Feature Engineering

```python
# Label Encoding for Ordinal Data

le = LabelEncoder()
label = ['Item_Fat_Content', 'Outlet_Type', 'Outlet_Location_Type', 'Outlet_Size']

for i in label:
    data[i] = le.fit_transform(data[i])
data.head()
```

```python
# One-Hot encoding (OHE) for Nominal Data

# Columns for applying One-Hot encoding
cols = ['Item_Type_Combined']

# Apply one-hot encoder
OneHot_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
data_onehot = pd.DataFrame(OneHot_encoder.fit_transform(data[cols])).astype('int64')

# get feature columns
data_onehot.columns = OneHot_encoder.get_feature_names_out(cols)

# One-hot encoding removed index; put it back
data_onehot.index = data.index

# Add one-hot encoded columns to our main df new name: tr_fe, te_fe (means feature engeenired)
data_feature = pd.concat([data, data_onehot], axis=1)
data_feature.head()
```

```python
# Drop irrlevent Columns

data_feature = data_feature.drop(['Item_Identifier','Item_Type_Combined', 'Outlet_Identifier','Outlet_Age',
                     'Item_Fat_Content','Item_Visibility', 'Item_Type','Item_Type_Combined',
                     'Outlet_Establishment_Year','Item_Type'], axis=1)
data_feature.head()

# Divide Data into train and test

train = data_feature.loc[data_feature['source']=="train"]
test = data_feature.loc[data_feature['source']=="test"]

train = train.drop('source', axis=1)
test = test.drop(['source',  'Item_Outlet_Sales'], axis=1)

# Check Datasets

print('\nTrain Dataset for Model Buidling: \n')
print(train.info(verbose=True, show_counts=True))

print('\nTest Dataset for Model Buidling: \n')
print(test.info(verbose=True, show_counts=True))

train.head()
```

## Train and Test split

```python
y = train['Item_Outlet_Sales']
X = train.drop('Item_Outlet_Sales', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 20)

# Cross Validation Score after model completion

def cross_val(model, X, y, cv):
    scores = CVS(model, X, y, cv=cv)
    print(f'{model} Scores:')
    for i in scores:
        print(round(i,2))
    print(f'Average {model} score: {round(scores.mean(),4)}')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Create a pipeline to standardize the data before applying linear regression
model = make_pipeline(StandardScaler(), LinearRegression())

# model fit
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)
```

```python
# Score Metrics for Regression:

print(f" Mean Absolute Error: {MAE(y_test, y_pred)}\n")
print(f" Squared Mean Squared Error: {np.sqrt(MSE(y_test, y_pred))}\n")
print(f" R^2 Score: {R2(y_test, y_pred)}\n")
```

```python
# Cross Validation Score check
cross_val(LinearRegression(), X, y, 5)
```
